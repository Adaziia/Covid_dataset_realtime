```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv")
```

```python
unique_locations=df["location"].nunique()
print(unique_locations)
```

```python
max_continent=df["continent"].value_counts().idxmax()
print(max_continent)
```

```python
Max_value_total_cases=df["total_cases"].max()
print(Max_value_total_cases)
```

```python
Mean_value_total_cases=df["total_cases"].sum()/len(df["total_cases"])
print(Mean_value_total_cases)
```

```python
df["total_deaths"].describe().round(2)
```

```python
df.groupby("continent")['gdp_per_capita'].mean().reset_index()
```

```python
df.groupby("continent")["human_development_index"].mean().reset_index()
```

```python
df=df[["continent","location","date","total_cases","total_deaths","gdp_per_capita","human_development_index"]]
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=["continent"])
```

```python
df.fillna(0)
```

```python
df["date"]=pd.to_datetime(df["date"])
```

```python
df.head()
```

```python
df["month"]=pd.DatetimeIndex(df["date"]).month
df
```

```python
df_groupby=df.groupby("continent").max().reset_index()
print(df_groupby)
```

```python
df_groupby['total_deaths to total_cases_ratio']=df_groupby["total_deaths"]/df_groupby["total_cases"]
```

```python
df_groupby["total_deaths to total_cases_ratio"]*100
```

```python
snr.histplot(df_groupby["gdp_per_capita"],bins=20)
plt.title("GPD_PER_CAPITA")
             
```

```python
snr.jointplot(x="gdp_per_capita",y="total_cases",hue="gdp_per_capita",data=df_groupby,kind="scatter")
```

```python
snr.pairplot(df_groupby,hue="continent")
```

```python
snr.catplot(x="continent",y="total_cases",data=df_groupby,kind='bar')
```

```python
df_groupby.to_csv("Covid_Data.csv")
```
